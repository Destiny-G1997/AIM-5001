```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
filename = r'C:\Users\guozi\Documents\GitHub\AIM-5001/Fatalities.csv'
tf = pd.read_csv(filename)# reading the csv file
#Drop the useless data
tf_useless = ['emppop','income', 'beertax','baptist', 'mormon', 'milestot', 'unempus',
              'emppopus', 'gsp']
tf.drop(labels = tf_useless, axis = 1,inplace = True)

tf.head()
```

## Exploratory Data Analysis

The **spirits** data is about spirits consumption from 1982-1988 in US.

```python
data = tf[['spirits']]# get the data of spirits consumption from 1982-1988 in US
plt.plot(data)
plt.title('Plot of spirits consumption')
plt.xlabel(' ')
plt.show()# show the data in Matplotlib.plot
data.describe()
```

**Conclusion:** The above plot and descriptive statistics show that the overall distribution of spirits consumption from 1982-1988 in different states with a minimum of 0.79, and a max of 4.9. That tells us the there should be outliers. Because, the max is 4.9, the 3rd quartile is 2.01. So every data over 2.01 could be an outlier. And we can know that the spirits consumptions are very different in different states.

The **Drinkage** is the minimum legal drinking age from 1982-1988 in US.

The 'fatal' column means number of vehicle fatalities in Alabam from 1982-1988.

```python
data3 = tf[['fatal','year']][(tf['state'] == 'al')] # get the data of number of vehicle fatalities in Alabam from 1982-1988.
plt.plot(data3['year'],data3['fatal'])
plt.title('Plot of number of vehicle fatalities in different states from 1982-1988')
plt.xlabel(' ')
plt.show()# show the data of number of vehicle fatalities in Alabam from 1982-1988 by year to year
sns.lineplot(x = data3['year'], y = data3['fatal']) # use the seaborn to show the data
data3['fatal'].describe()
```

**Conclusion:** The above plot and seaborn lineplot show us that the number of vehicle fatalities in Alabam from 1982-1988 with a mean of 971, a min of 839, and a max of 1110. And we can know the tendency of this data is on the rise in these 7 years.

## Inference

**Research question:** From 1982-1988 what's the percent of youngdirver in traffic fatalities changes in Alabam?

- To get a sub set data use for the research question.
- The state column is the states of the US. The youngdrivers means percent of drivers aged 15–24.

```python
tf1 = tf[['state','year','youngdrivers']]#get the data of the percent of youngdirver in traffic fatalities changes in Alabam
#from 1982-1988
yr = tf1[tf1['state']=='al']
print(type(yr))
yr
```

```python
plt.plot(yr['year'],yr['youngdrivers'])
plt.title('youngdriver changes from 1982-1988 in Alabam')
plt.xlabel(' ')
plt.show()#show the percent of youngdirver in traffic fatalities changes in Alabam year by year
sns.lineplot(x = yr['year'], y = yr['youngdrivers'])#use the seaborn linplot to show the data
yr['youngdrivers'].describe()
```

**Result:** The above matplotlib plot, seaborn lineplot, and descriptive statistics show that 'youngdirver' changes in Alabam from 1982-1988 with a mean of 0.213, a minimum of 0.210, and a max of 0.218. And from the plot, we can know that the percent of young drivers in traffic fatalities didn't change too much from 1982-1988. But starting from 1985 the data is on the rise.
