```python
subdata['gill-size'].drop_duplicates()
```

```python
gillsizeDict = {
    'n': 0,
    'b': 1
}
subdata['gill-size'] = subdata['gill-size'].map(gillsizeDict)
subdata
```

```python
test = subdata
ddf12 = ddf.join(ddf2)
ddf123 = ddf12.join(ddf3)
train = ddf123.join(ddf4)
test
```

## Conclusion:  
According to the OLS summary, there are 8124 of observations. And this summary is analyzed by using the least squares method. The degrees of freedom of the residuals of odor is 8115. And there are 8 Df models. The **R-squared** is the coefficient of determination.It's a statistical measure of how well the regression line approximates the real data points. The **R-squared** of odor is 1.00. The **Adj.R-squared** is the above value adjusted based on the number of observations and the degrees-of-freedom of the residuals. The **Adj.R-squared** of odor is 1.00. So that we can know the explanatory ability of the variables and the data fitting of this model is perfect. The **F-statistic** is a measure how significant the fit is. The **F-statistic** of odor is 4.043e+30.
