```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.random.randn(50).cumsum(),'k--');#k-- = black line
```

```python
plt.hist(np.random.randn(50));
```

```python
plt.boxplot(np.random.randn(50));
```

```python
plt.figure(1)#define matplotlib figures
plt.hist(np.random.randn(100))
plt.title('Histogram of 100 Samples From Normal Distribution')
plt.locator_params(nbins = 10, axis = 'x')#define the number of bins for the histogram + which plot axis they will
#be applied to
plt.show()
```

```python
#side-by-side boxplot and histogram of one set of data
data = np.random.randn(100)
plt.figure(1)
#the '121' value passed to the function says:"I am defining a 1x2"arrangement of subplots(as 
#indicatdby the first 2 digits '12'), and this is the specification for the plot that will
#appear on the lefthand side (as indicated by the last digit'1')"
plt.subplot(121)
plt.boxplot(data)
plt.title('Boxplot of Data')
plt.xlabel(' ')
#--------------------------------------------------------------------------------------------
plt.subplot(122)
plt.hist(data)
plt.title('Histogram of data')
plt.locator_params(nbins = 10, axis = 'x')
plt.show()
```

```python
#define color, markers and linestyle
plt.plot(np.random.randn(30).cumsum(),color = 'g', marker = 'o', linestyle = 'dashed');
```

```python
plt.plot(np.random.randn(30).cumsum(), 'go--');
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0,250,500,750,100])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 30, 
                            fontsize = 'small')
ax.set_title('Example from Page 267-8 in pfDA Text')
ax.set_xlabel('Stages');
```

```python
fig = plt.figure();
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k',label = 'one')
ax.plot(np.random.randn(1000).cumsum(), 'k--',label = 'two')
ax.plot(np.random.randn(1000).cumsum(), 'k.',label = 'three')
ax.legend(loc = 'best');
```

# Seaborn

```python
import seaborn as sns
sns.lineplot(data = np.random.randn(50).cumsum())
```

```python
sns.distplot(np.random.randn(1000), bins = 10)
```

```python
sns.boxplot(np.random.randn(50), orient = 'v')
```

```python
data = np.random.randn(1000)
plt.figure(1)
plt.subplot(121)
sns.boxplot(data, orient = 'v')
plt.title('Boxplot of Data')
plt.xlabel(' ')
#----------------------------------------------------------------------------------
plt.subplot(122)
sns.distplot(data, bins = 10)
plt.title('Histogram of Data')
plt.show();
```
