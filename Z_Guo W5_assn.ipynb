# Q1:From 'IAH' witch Dest has the farthest distance?

```python
flight1 = flight['Distance'].max()
print(flight1)# get the farthest distance
flight[(flight['Distance']==3904)][['Origin','Dest','Distance']]
```

# Q2:What's the all arrtime lineplot in February 1st afternoon?

```python
gplot = flight[(flight['ArrTime'] > 1200)&(flight['Month'] ==2)&(flight['DayofMonth'] ==1)]
gplot.plot(x='DayofMonth',y='ArrTime')# Create the line plot
```

I created a new df called 'gplot'. Inputed the value of arrtime in February 1st afternoon. And then use the .plot function to make this lineplot. The arrtime in February 1st afternoon is from 14:00-21:00.

# Q3:Which day in February had the longgest arrdelay?

```python
delay = flight.loc[:,['ArrDelay']]
print(delay.max())# get the max value of ArrDelay
flight[flight['ArrDelay']==775.0][['Month','DayofMonth']]
```
